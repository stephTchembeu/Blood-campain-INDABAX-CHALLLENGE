### We want to precise that here we have handle some errors in the date.

what we have done:
- we have wrote a function to convert the element of this column into datetime
- our function change the dates which were in mm/dd/yy into dd/mm/yy and convert.
- we have set by hand to the first of january 2019 the following observtion which has the birthdate insted to the fill date.

```python
def preprocess_date(date_str):
    try:
        # we attempt to convert to datetime
        date = pd.to_datetime(date_str)
    except ValueError:
        # If conversion fails, we assume the date is in the format 'm/d/yyyy'
        parts = date_str.split('/') # we split dates using / as separator 
        if len(parts) == 3 and int(parts[2]) < 1000: 
            # change the year in the good format with the year at 2019
            date_str = f"{parts[1]}/{parts[0]}/2019"
        date = pd.to_datetime(date_str)
    return date
```

```python
# Apply the preprocessing function
sheet_2019['Date de remplissage de la fiche'] = sheet_2019['Date de remplissage de la fiche'].apply(preprocess_date)
```

```python
# Next column
def preprocess_birthdate(date_str):
    try:
        # Attempt to convert to datetime
        date = pd.to_datetime(date_str)

    except ValueError:
        # If conversion fails, assume the date is in the format 'm/d/yyyy'
        parts = date_str.split('/')
        if len(parts) == 3 and int(parts[2]) < 1000:
            # Convert to 'mm/dd/yyyy' format
            if int(parts[2]) < 10:
                parts[2] = str(2000 + int(parts[2]))
            elif int(parts[2]) > 10 and int(parts[2]) <= 99:
                parts[2] = str(1900 + int(parts[2]))

            date_str = f"{parts[0]}/{parts[1]}/{parts[2]}"
        date = pd.to_datetime(date_str)
    return date

sheet_2019["Date de naissance"] = sheet_2019["Date de naissance"].apply(preprocess_birthdate)
```

```python
# we assume that people with year "2089, 2092" wanted to put 1989 and 1992 so we change their year
def fix_birthdate_year(date):
    if date.year > 2019:
        if date.year == 2089:
            date = date.replace(year=1989)
        elif date.year == 2092:
            date = date.replace(year=1992)
    return date

sheet_2019["Date de naissance"] = sheet_2019["Date de naissance"].apply(fix_birthdate_year)
```

```python
def infers_education_level(profession):
    profession = profession.lower().strip()
    # Professions requiring no formal education
    no_education = {"manœuvre", "ouvrier", "agriculteur", "pêcheur"}
    # Professions accessible at student level (basic training or apprenticeship)
    student_level = {"apprenti", "assistant", "aide", "vendeur", "serveur", "chauffeur"}
    # Professions requiring secondary education (CAP, Bac Pro, etc.)
    secondary_level = {
        "soudeur", "électricien", "mécanicien", "coiffeur", "plombier", "maçon", 
        "esthéticien", "technicien", "chauffagiste", "menuisier", "couvreur",
        "chaudronnier", "peintre", "carrossier", "boulanger", "pâtissier", "boucher"
    }
    # Professions requiring university education
    university_level = { "etudiant (e)",
        "ingénieur", "médecin", "avocat", "pharmacien", "architecte", "chercheur", 
        "professeur", "économiste", "développeur", "informaticien", "juriste", "comptable",
        "psychologue", "notaire", "dentiste", "vétérinaire", "biologiste", "scientifique"
    }
    
    # assigning education levels
    if any(word in profession for word in no_education):
        return "pas précisé"
    elif any(word in profession for word in student_level):
        return "élève"
    elif any(word in profession for word in secondary_level):
        return "secondaire"
    elif any(word in profession for word in university_level):
        return "universitaire"
    else:
        return "pas précisé"  # If the profession is difficult to classify
```

```python

def infer_education_level(row):
    if row["Niveau d\'etude"] in ["pas précisé", "aucun"]:
        return infers_education_level(row["Profession"])
    return row["Niveau d\'etude"]  

sheet_2019["Niveau d\'etude"] = sheet_2019.apply(infer_education_level,axis=1)
```
